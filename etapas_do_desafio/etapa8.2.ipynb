# Número de apps que não tem restrição de idade (Everyone): (gráfico)
- Segue a mesma linha de raciocínio do outro código, mas com a criação de um gráfico de linhas e a classificação dos apps livres por categoria, nesse caso os apps da categoria família são os mais livres.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('googleplaystore_clean.csv')

todos_apps = df[df['Content Rating'] == 'Everyone']

todos_apps_by_category = todos_apps['Category'].value_counts()

plt.figure(figsize=(12, 6))
todos_apps_by_category.plot(kind='line', marker='o')
plt.title("Número de aplicativos classificados como 'Everyone' por categoria")
plt.xlabel("Categoria")
plt.ylabel("Número de aplicativos")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()

```
