# Categorias de APPS  existentes no datase: 
- Pedia para mostrar de acordo com a frequência que elas apareciam e montar um gráfico de pizza. Eram muitas categorias, então criei dois gráficos. Nesse caso, foi necessário o uso do matplotlib, seguindo os mesmos passos dos outros: fazendo a leitura do arquivo CSV, contando a frequência das categorias e calculando a porcentagem em que elas aparecem. As categorias foram separadas em dois gráficos: um para as categorias com frequência maior que 2,4% e outro para as categorias com frequência menor, para ficar mais organizado. Depois, foram criados dois gráficos de pizza.t

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('googleplaystore_clean.csv')

# Faz a contagem:
category_counts = df['Category'].value_counts()

# Calcula a porcentagem:
category_percentage = category_counts / category_counts.sum() * 100

# Criei dois graficos, pois um só ficou muito feio e não dava para entender. Nisso criei um chamado outros para dividir.
common_categories = category_percentage[category_percentage >= 2.4]
uncommon_categories = category_percentage[category_percentage < 2.4]

plt.figure(figsize=(10, 6))
plt.pie(common_categories, labels=common_categories.index, autopct='%1.1f%%', startangle=140)
plt.title('Categorias de Aplicativos:')
plt.axis('equal') 
plt.show()

plt.figure(figsize=(10, 6))
plt.pie(uncommon_categories, labels=uncommon_categories.index, autopct='%1.1f%%', startangle=140)
plt.title('Outros:')
plt.axis('equal') 
plt.show()

```
