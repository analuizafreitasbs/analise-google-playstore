# Top 5 apps por número de instalação:
- O código carrega o arquivo CSV, converte os dados da coluna "install" para dados numéricos e seleciona os 5 aplicativos com mais de 1.000.000.000 de instalações. No total, existem 49 aplicativos com mais de 1.000.000.000 de instalações, mas apenas 5 são necessários. Em seguida, é criado um gráfico de barras para visualizar esses 5 aplicativos. No caso os aplicativos que saíram foi: Google play books, Messenger, Whatszap, Chrome e gmail.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('googleplaystore_clean.csv')

df = df[df['Installs'] != 'Free']

# conversão: 
df['Installs'] = df['Installs'].str.replace('+', '')
df['Installs'] = df['Installs'].str.replace(',', '')
df['Installs'] = df['Installs'].astype(int)

top_apps = df[df['Installs'] >= 1000000000].nlargest(5, 'Installs')
# existia lá 49 apps com mais de 1,000,000,000 de installs, então teria várias possibilidades essa tabela. No caso 1B era o máximo.

plt.figure(figsize=(10, 6))
plt.bar(top_apps['App'], top_apps['Installs'], color='skyblue')
plt.title('Top 5 Apps por Número de Instalações (1 Bilhão ou Mais)')
plt.xlabel('Aplicativo')
plt.ylabel('Número de Instalações')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


```
