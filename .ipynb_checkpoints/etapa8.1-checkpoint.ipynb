# Top 10 apps com menos Reviews: (gráfico)
- Segue a mesma linha de raciocínio do outro código, mas com a criação de um gráfico de dispersão.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('googleplaystore_clean.csv')

# remove as duplicadas:
df = df.drop_duplicates(subset='App', keep='first')

# faz a conversão de reviews para tipo númerico:
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')

top_10_menos_reviews = df[['App', 'Reviews']].nsmallest(10, 'Reviews')

plt.figure(figsize=(10, 6))
plt.scatter(top_10_menos_reviews['App'], top_10_menos_reviews['Reviews'], color='skyblue')
plt.title('Top 10 Apps com Menos Reviews (Installs > 0)')
plt.xlabel('Aplicativo')
plt.ylabel('Número de Reviews')
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```
